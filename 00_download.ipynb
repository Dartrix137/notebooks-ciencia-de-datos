```python
#Importación de librerías
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', 90)
```

```python
# CWD = Current Working Directory
cwd = os.getcwd()
raw_dir = cwd + "/../data/raw/"
processed_dir = cwd + "/../data/processed/"
```

version 1: utilizando `os.listdir()`

```python
# obtener todos los nombres de los archivos XLSX que necesitamos
files = []
paths = os.listdir(raw_dir)
for path in paths:
    p = [ raw_dir + path + '/' + file for file in os.listdir(raw_dir + path) ]
    files.extend(p)
files
```

```python
#Creacion de dataframes para cada archivo
rs_2019_1 = pd.read_csv(files[0], sep='|')
rs_2019_2 = pd.read_csv(files[1], sep='¬')
rs_2020_1 = pd.read_csv(files[2], sep='¬')
rs_2020_2 = pd.read_csv(files[3], sep='¬')
rs_2021_2 = pd.read_csv(files[4], sep='¬')
rs_2022_1 = pd.read_csv(files[5], sep='¬')
rs_2022_2 = pd.read_csv(files[6], sep='¬')
```

```python
rs_2019_1
```

```python
#Se filtra cada dataframe para registros del departamento del Valle
rs_2019_1 = rs_2019_1[rs_2019_1.estu_depto_reside == "VALLE"]
rs_2019_2 = rs_2019_2[rs_2019_2.ESTU_DEPTO_RESIDE == "VALLE"]
rs_2020_1 = rs_2020_1[rs_2020_1.ESTU_DEPTO_RESIDE == "VALLE"]
rs_2020_2 = rs_2020_2[rs_2020_2.ESTU_DEPTO_RESIDE == "VALLE"]
rs_2021_2 = rs_2021_2[rs_2021_2.ESTU_DEPTO_RESIDE == "VALLE"]
rs_2022_1 = rs_2022_1[rs_2022_1.ESTU_DEPTO_RESIDE == "VALLE"]
rs_2022_2 = rs_2022_2[rs_2022_2.ESTU_DEPTO_RESIDE == "VALLE"]
```

```python
#Se eliminan algunas columnas para que todos los dataframes tengan las mismas columnas
rs_2019_1= rs_2019_1.drop(columns=['estu_generacione'])
rs_2019_2 = rs_2019_2.drop(columns=['ESTU_ETNIA', 'ESTU_GENERACION-E'])
rs_2020_1 = rs_2020_1.drop(columns=['ESTU_GENERACION-E'])
rs_2020_2 = rs_2020_2.drop(columns=['ESTU_GENERACION-E'])
rs_2021_2 = rs_2021_2.drop(columns=['ESTU_GENERACION-E', 'PERCENTIL_ESPECIAL_GLOBAL'])
rs_2022_1 = rs_2022_1.drop(columns=['PERCENTIL_ESPECIAL_GLOBAL'])
rs_2022_2 = rs_2022_2.drop(columns=['PERCENTIL_ESPECIAL_GLOBAL'])
```

```python
#Se añade una columna a cada data set para espedificar el semestre al que pertenecen los datos
rs_2019_1["SEMESTRE"]= "2019-1"
rs_2019_2["SEMESTRE"] = "2019-2"
rs_2020_1["SEMESTRE"] = "2020-1"
rs_2020_2["SEMESTRE"] = "2020-2"
rs_2021_2["SEMESTRE"] = "2021-2"
rs_2022_1["SEMESTRE"] = "2022-1"
rs_2022_2["SEMESTRE"] = "2022-2"
```

```python
#Se reacomodan las columnas del dataframe 2019_1 para coincidir con las de 2022_2
columnas1=list(rs_2022_2.columns)
columnas2=list(rs_2019_1.columns)
for i, col in enumerate(columnas1):
  if(col==columnas2[i].upper()):
    pass
  else:
    for j, col2 in enumerate(columnas2):
      if(col2.upper()==col):
        x, y=columnas2.index(columnas2[i]), columnas2.index(columnas2[j])
        columnas2[y], columnas2[x]= columnas2[x], columnas2[y]
        rs_2019_1=rs_2019_1[columnas2]
```
